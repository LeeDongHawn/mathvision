```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D # <--- This is important for 3d plotting 
```

```python
# f(x,y) = (x+y)(xy+xy^2)
def f(x, y):
    return (x+y)*(x*y+x*y*y)
```

```python
# f(x,y) = (x+y)(xy+xy^2) 함수의 그래프 그리기
from matplotlib.lines import Line2D   
fig = plt.figure(figsize=(20, 15))
ax = fig.gca(projection='3d')

a = np.arange(-1, 1.6, 0.1)
b = np.arange(-1.2, 0.3, 0.1)

a, b = np.meshgrid(a, b)
c = f(a,b)
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.3, 
                       linewidth=0, antialiased=False, cmap='rainbow')


   
ax.scatter(1,0, marker='x', s=100)

ax.scatter(0,1, marker='s', s=100)

x = [1,0]
y = [0,1]
ax.plot(x,y)


plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# fx(x에 대해 편미분)
def fx(x, y):
    return x*y**2 + x*y + (x + y)*(y**2 + y)
# fy(y에 대해 편미분)
def fy(x, y):
    return x*y**2 + x*y + (x + y)*(2*x*y + x)
```

```python
import sympy
import numpy as np
import matplotlib.pyplot as plt

x, y = sympy.symbols("x y")

print(f)

# Hessian Matrix
f = (x+y)*(x*y+x*y*y)
fx = sympy.diff(f, x)
fy = sympy.diff(f, y)
h_x = sympy.diff(fx,x)
h_xy = sympy.diff(fx,y)
h_yx = sympy.diff(fy,x)
h_yy = sympy.diff(fy,y)

print(h_x)
print(h_xy)
print(h_yx)
print(h_yy)
print(fx)
print(fy)

# 임계점(critical point)
print(fx.evalf(subs={x:0, y:0}))
print(fy.evalf(subs={x:0, y:0}))
print(fx.evalf(subs={x:0, y:-1}))
print(fy.evalf(subs={x:0, y:-1}))
print(fx.evalf(subs={x:1, y:-1}))
print(fy.evalf(subs={x:1, y:-1}))
print(fx.evalf(subs={x:3/8, y:-3/4}))
print(fy.evalf(subs={x:3/8, y:-3/4}))
```

```python
# Hessian 판별식 D
hessian = [h_x, h_xy, h_yx, h_yy]
print(hessian)

xx1 = hessian[0].evalf(subs={x:0, y:0})
xy1 = hessian[1].evalf(subs={x:0, y:0})
yx1 = hessian[1].evalf(subs={x:0, y:0})
yy1 = hessian[3].evalf(subs={x:0, y:0})

hessian1 = [xx1, xy1, yx1, yy1]
print(hessian1)

xx2 = hessian[0].evalf(subs={x:0, y:-1})
xy2 = hessian[1].evalf(subs={x:0, y:-1})
yx2 = hessian[2].evalf(subs={x:0, y:-1})
yy2 = hessian[3].evalf(subs={x:0, y:-1})

hessian2 = [xx2, xy2, yx2, yy2]
print(hessian2)

xx3 = hessian[0].evalf(subs={x:1, y:-1})
xy3 = hessian[1].evalf(subs={x:1, y:-1})
yx3 = hessian[2].evalf(subs={x:1, y:-1})
yy3 = hessian[3].evalf(subs={x:1, y:-1})

hessian3 = [xx3, xy3, yx3, yy3]
print(hessian3)

xx4 = hessian[0].evalf(subs={x:3/8, y:-3/4})
xy4 = hessian[1].evalf(subs={x:3/8, y:-3/4})
yx4 = hessian[2].evalf(subs={x:3/8, y:-3/4})
yy4 = hessian[3].evalf(subs={x:3/8, y:-3/4})

hessian4 = [xx4, xy4, yx4, yy4]
print(hessian4)
```
