```python
import sympy
import numpy as np
import math

x, y = sympy.symbols("x y")

# gradient (1)
f = sympy.sin(x+y-1)+(x-y-1)*(x-y-1)-1.5*x+2.5*y+1
fx = sympy.diff(f, x)
fy = sympy.diff(f, y)

print(fx)
print(fy)
```

```python
# f(x,y) = sin(x+y-1)+(x-y-1)^2-1.5x+2.5y+1
def f(x, y):
    return np.sin(x+y-1)+(x-y-1)*(x-y-1)-1.5*x+2.5*y+1
```

```python
# gradient descent method
# lambda, iterator  
lam = 0.01
iteration = 100

#     
sol = np.zeros((iteration,2))
sol[0][0] = -1
sol[0][1] = -1

# x(k+1) = x(k) - lam * f'(x(k))
# f' "gradient"
for i in range(1,iteration):
    sol[i][0]=sol[i-1][0] - lam*fx.evalf(subs={x:sol[i-1][0], y:sol[i-1][1]})
    sol[i][1]=sol[i-1][1] - lam*fy.evalf(subs={x:sol[i-1][0], y:sol[i-1][1]})
    

#   
fig = plt.figure(figsize=(20,15))
ax = fig.gca(projection='3d')
a = np.arange(-1, 5, 1)
b = np.arange(-3, 4, 1)
a, b = np.meshgrid(a, b)
c = f(a,b)

surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.3, 
                       linewidth=0, antialiased=False, cmap='rainbow')

#   
for i in range(0,iteration):
    
    x = sol[i][0]
    y = sol[i][1]
    z = f(x,y)
    if(i==0):
        ax.scatter(x, y, z, s=50, marker='s',c='b')
        print("first_z : ",z)
    if(i==iteration-1):
        ax.scatter(x, y, z, s=50, marker='s',c='r')
        print("last_z : ",z)
    else:
        ax.scatter(x, y, z, s=50, marker='x',c='g')
        print("middle_z : ",z)

plt.show()
```

```python
import sympy
import numpy as np
import math

x, y = sympy.symbols("x y")

# hessian (2)
f = sympy.sin(x+y-1)+(x-y-1)*(x-y-1)-1.5*x+2.5*y+1
fx = sympy.diff(f, x)
fy = sympy.diff(f, y)
h_xx = sympy.diff(fx, x)
h_xy = sympy.diff(fx, y)
h_yx = sympy.diff(fy, x)
h_yy = sympy.diff(fy, y)

print(h_xx)
print(h_xy)
print(h_yx)
print(h_yy)
```

```python
# f(x,y) = sin(x+y-1)+(x-y-1)^2-1.5x+2.5y+1
def f(x, y):
    return np.sin(x+y-1)+(x-y-1)*(x-y-1)-1.5*x+2.5*y+1
```

```python
# newton's method
# iterator  ( lambda  )
iteration = 10

#     
sol1 = np.zeros((iteration,2))
sol1[0][0] = -1
sol1[0][1] = -1

# h "hessian" f "gradient"
for i in range(1,iteration):
    fxx = fx.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
    fyy = fy.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
    hxx = h_xx.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
    hxy = h_xy.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
    hyx = h_yx.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
    hyy = h_yy.evalf(subs={x:sol1[i-1][0], y:sol1[i-1][1]})
#     print(hxx)
#     print(fxx)

    # 2x2 numpy array .
    hessian = np.array([[hxx,hxy],[hyx,hyy]])
    # numpy array float   .
    mat_h = np.array([[hxx,hxy],[hyx,hyy]]).astype(np.float64)

    # 
    # u : left singular vector
    # w : singular value
    # vt : right singular value
    u, w, vt = np.linalg.svd(mat_h)
    # w   saddle-free  .
#     w = np.abs(w)
    #    .
    w = np.diag(w)
    # u * w * vt   hessian  .
    result_hesian = np.matmul(np.matmul(u, w), vt)
    
    #  hessian   .
    result_hesian_inv = np.linalg.inv(result_hesian)
#     print(result_hesian_inv)
    
    #     .
    result_xx = result_hesian_inv[0,0]
    result_xy = result_hesian_inv[0,1]
    result_yx = result_hesian_inv[1,0]
    result_yy = result_hesian_inv[1,1]
    
    # x(k+1) = x(k) - (H(x(k))^-1 * f'(x(k))  
    sol1[i][0]= sol1[i-1][0] - (result_xx*fxx + result_xy*fyy)
    sol1[i][1]= sol1[i-1][1] - (result_yx*fxx + result_yy*fyy)

      
#   
fig = plt.figure(figsize=(20,15))
ax = fig.gca(projection='3d')
a1 = np.arange(-1, 5, 1)
b1 = np.arange(-3, 4, 1)
a1, b1 = np.meshgrid(a1, b1)
c1 = f(a1,b1)

surf1 = ax.plot_surface(a1, b1, c1, rstride=1, cstride=1, alpha=0.3, 
                       linewidth=0, antialiased=False, cmap='rainbow')

#   ()
for i in range(0,iteration):
    
    x1 = sol1[i][0]
    y1 = sol1[i][1]
    z1 = f(x1,y1)
    if(i==0):
        ax.scatter(x1, y1, z1, s=50, marker='s',c='b')
        print("first_z : ",z1)
    if(i==iteration-1):
        ax.scatter(x1, y1, z1, s=50, marker='s',c='r')
        print("last_z : ",z1)
    else:
        ax.scatter(x1, y1, z1, s=50, marker='x',c='g')
        print("middle_z : ",z1)

plt.show()
```
